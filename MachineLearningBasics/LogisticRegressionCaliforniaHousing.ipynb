```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_data = pd.read_csv('datasets/housing.csv') # Previously explored Dataset for California Housing
housing_data.sample(10)
```

```python
housing_data.dropna(inplace=True)
```

```python
housing_data.describe().transpose()
```

```python
housing_data.loc[housing_data['median_house_value'] == 500001].count()
```

```python
housing_data = housing_data.drop(housing_data.loc[housing_data['median_house_value'] == 500001].index)
housing_data.describe()
```

```python
housing_data.shape
```

```python
housing_data['ocean_proximity'].value_counts().sort_values()
```

```python
housing_data = pd.get_dummies(housing_data,columns=['ocean_proximity'])
housing_data.sample(5)
```

```python
# Converting meadian house value to 0 or 1 (< meadian , > median) for classification problem
median = housing_data['median_house_value'].median()
```

```python
housing_data['above_median'] = housing_data['median_house_value'] > median
```

```python
housing_data.drop(['median_house_value'], axis=1, inplace=True)
```

```python
housing_data.sample(5)
```

### Preparing and Budiling Classification Model

```python
X = housing_data.drop(['above_median'], axis=1)
Y = housing_data['above_median']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear').fit(x_train, y_train)
print('Training Score:', lr.score(x_train, y_train))
```

```python
y_pred = lr.predict(x_test)

from sklearn.metrics import accuracy_score
print('Testing Score: ', accuracy_score(y_test, y_pred))
```
