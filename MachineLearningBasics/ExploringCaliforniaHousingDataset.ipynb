### Visualizing Relationships

```python
# total_rooms vs median_house_value

fig, ax = plt.subplots(figsize=(10,7))
plt.scatter(housing_data['total_rooms'], housing_data['median_house_value'])
plt.xlabel('total_rooms')
plt.ylabel('median_house_value')
plt.show()
```

```python
# housing_median_age vs median_house_value
fig, ax = plt.subplots(figsize=(10,7))
plt.scatter(housing_data['housing_median_age'], housing_data['median_house_value'])
plt.xlabel('Housing Median Age')
plt.ylabel('Median House Value')
plt.show()
```

```python
# median_income vs median_house_value
fig, ax = plt.subplots(figsize=(10,7))
plt.scatter(housing_data['median_income'], housing_data['median_house_value'])
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
housing_data_corr = housing_data.corr()
housing_data_corr
```

```python
fig, ax = plt.subplots(figsize=(10,7))
sns.heatmap(housing_data_corr, annot=True)
```
